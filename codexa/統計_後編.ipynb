Chapter7　共分散　相関係数

````python
import numpy as np
````

````python
import pandas as pd
````

````python
import matplotlib.pyplot as plt
````

````python
x = np.array([1,3,4,6])
````

````python
y = np.array([1,2,5,8])
````

````python
x_std = x.std()
````

````python
y_std = y.std()
````

````python
x_std
````

````python
y_std
````

````python
cov = np.cov(x,y,ddof=0)[0][1]
````

````python
cov
````

````python
r = cov / (x_std * y_std)
````

````python
r
````

````python
#おまけ
mean = np.array([0,0]) #平均を指定。
````

````python
cov01 = np.array([
    [1,0.1],
    [0.1,1]]) #共分散行列を指定。0.1のところを変えればxとyの共分散の値を変えられる。
````

````python
x , y = np.random.multivariate_normal(mean,cov01,5000).T #とりあえず5000個生成
````

````python
plt.plot(x,y,'x') #プロットする
````

````python
plt.title('covariance = 0.1')
````

````python
plt.axis('equal')
````

````python
plt.show()
````

````python
cov03 = np.array([
    [1,0.3],
    [0.3,1]]) #共分散行列を指定。0.1のところを変えればxとyの共分散の値を変えられる。
x , y = np.random.multivariate_normal(mean,cov03,5000).T #とりあえず5000個生成
plt.plot(x,y,'x')
plt.title('covariance = 0.3')
plt.axis('equal')
plt.show()
````

````python
cov05 = np.array([
    [1,0.5],
    [0.5,1]]) #共分散行列を指定。0.1のところを変えればxとyの共分散の値を変えられる。
x , y = np.random.multivariate_normal(mean,cov05,5000).T #とりあえず5000個生成
plt.plot(x,y,'x')
plt.title('covariance = 0.5')
plt.axis('equal')
plt.show()
````

````python
cov07 = np.array([
    [1,0.7],
    [0.7,1]
])
x,y = np.random.multivariate_normal(mean,cov07,5000).T
plt.plot(x,y,'x')
plt.title('covariance = 0.7')
plt.axis('equal')
plt.show()
````

````python
cov09 = np.array([
    [1,0.9],
    [0.9,1]
])
x,y = np.random.multivariate_normal(mean,cov09,5000).T
#plt.plot(x,y,'x')
plt.plot(x,y,'ro')
plt.title('covariance = 0.9')
plt.axis('equal')
plt.show()
````

````python
cov10 = np.array([
    [1,1],
    [1,1]
])
x,y = np.random.multivariate_normal(mean,cov10,5000).T
plt.plot(x,y,'rx')
plt.title('covariance = 1.0')
#plt.axis('equal') 
#xあるいはyの上限・下限を調整して同じ座標値の増分が同じ長さになるように調整します。円がちゃんと丸く表示されます。
plt.axis('scaled') 
#achieves the same result by changing the dimensions of the plot box instead of the axis data limits.
#上限・下限はそのままで　表示する領域を調整することで'equal'の場合と同じ効果を得ます。
plt.show()
````
